```python
import pandas as pd
Counsellors = pd.read_excel(r"/content/Counsellor.xlsx")
Faculty_Advisors = pd.read_excel(r"/content/FacultyAdvisor.xlsx")
Batch = pd.read_excel(r"/content/ClassRepresentative.xlsx")
```

```python
FA = {}
FA_title = []
Timetable_FA = {}
Days = ["Monday","Tuesday","Wednesday","Thursday","Friday"]
for i in Faculty_Advisors.columns:
    if(i!="Timestamp"):
        FA_title.append(i)
for i in range(len(Faculty_Advisors)):
    if Faculty_Advisors[FA_title[0]][i] not in FA:
        FA[Faculty_Advisors[FA_title[0]][i]]=[Faculty_Advisors[FA_title[1]][i],Faculty_Advisors[FA_title[2]][i]]
        Timetable_FA[Faculty_Advisors[FA_title[0]][i]] = {}
        for j in range(3,8):
            Timetable_FA[Faculty_Advisors[FA_title[0]][i]][Days[j-3]] = []
            Timetable_FA[Faculty_Advisors[FA_title[0]][i]][Days[j-3]].append(Faculty_Advisors[FA_title[j]][i])

Timetable_FA
```

```python
counsellor = {}
counsellor_title = []
Timetable_counsellor = {}
Days = ["Monday","Tuesday","Wednesday","Thursday","Friday"]
for i in Counsellors.columns:
    if(i!="Timestamp"):
        counsellor_title.append(i)
print(counsellor_title)
for i in range(len(Counsellors)):
    if Counsellors[counsellor_title[0]][i] not in counsellor:
        counsellor[Counsellors[counsellor_title[0]][i]]=[]
        Timetable_counsellor[Counsellors[counsellor_title[0]][i]]={}
        for j in range(1,1+len(Days)):
            Timetable_counsellor[Counsellors[counsellor_title[0]][i]][Days[j-1]] = []
            Timetable_counsellor[Counsellors[counsellor_title[0]][i]][Days[j-1]].append(Counsellors[counsellor_title[j]][i])
Timetable_counsellor
for i in Timetable_counsellor:
    for k in Days:
        Timetable_counsellor[i][k] = Timetable_counsellor[i][k][0].split(",")
        
Timetable_counsellor
```

```python
CR = {}
CR_title = []
Timetable_batch = {}
Days = ["Monday","Tuesday","Wednesday","Thursday","Friday"]
for i in Batch.columns:
    if(i!="Timestamp"):
        CR_title.append(i)
for i in range(len(Batch)):
    if Batch[CR_title[0]][i] not in CR:
        CR[Batch[CR_title[0]][i]]=[Batch[CR_title[1]][i],Batch[CR_title[2]][i]]
        Timetable_batch[Batch[CR_title[0]][i]] = {}
        for j in range(3,3+len(Days)):
            Timetable_batch[Batch[CR_title[0]][i]][Days[j-3]] = []
            Timetable_batch[Batch[CR_title[0]][i]][Days[j-3]].append(Batch[CR_title[j]][i])
CR
```

```python
ans=[]
for i in CR:
    for j in FA:
        temp=[]
        if CR[i][0]==FA[j][0] and CR[i][1]==FA[j][1]:
            # print(i,j)
            # print(CR[i][0],FA[j][0])
            # print(CR[i][1],FA[j][1])
            for l in range(3,3+len(Days)):
                if( Timetable_FA[j][Days[l-3]][0] == str(Timetable_FA[j][Days[l-3]][0]) and Timetable_batch[i][Days[l-3]][0]==str(Timetable_batch[i][Days[l-3]][0])):
                    
                    
                    overlap = (Timetable_FA[j][Days[l-3]][0])
                    overlap = overlap.split(",")
                    # print(i,j)
                    # print("List 1 is",overlap)
                    # print(Timetable_batch[i][Days[l-3]])
                    overlaptwo = (Timetable_batch[i][Days[l-3]][0])
                    overlaptwo = overlaptwo.split(",")
                    # Timetable_batch[i][Days[l-3]]=(Timetable_batch[i][Days[l-3]][0]).split(",")
                    # s=set((Timetable_batch[i][Days[l-3]]))
                    # s=set(Timetable_batch[i][Days[l-3]])
                    # print("List 2 is",overlaptwo)
                    overlap = [x for x in overlap if x in overlaptwo]
                    
                    if(len(overlap)>0):

                        
                        temp.append(Days[l-3])
                        temp.append(overlap)
            if len(overlap)>0:
               temp=[j]+temp;
               temp=[i]+temp;
               ans.append(temp)
print(ans)

data = []
for i in range(len(ans)):
    temp=[0,0,0,0,0,0,0,0]
    # print(len(ans[i]))
    temp[0]=CR[ans[i][0]][0]
    temp[1]=CR[ans[i][0]][1]
    temp[2]=ans[i][1]
    for j in range(2,len(ans[i]),2):
        if ans[i][j] in Days:
            for k in range(len(Days)):
                if Days[k]==ans[i][j]:
                    temp[3+k]=ans[i][j+1]
    data.append(temp)
#     print(data)
# print(data)

df=pd.DataFrame(data,columns=['Batch','Discipline','Faculty Advisor','Monday','Tuesday','Wednesday','Thursday','Friday'])
df




                

```

```python
writer = pd.ExcelWriter(r'/content/Faculty Adviser Schedule.xlsx')

# write dataframe to excel
 
df.to_excel(writer)
 
# save the excel
writer.save()
print("DataFrame is exported successfully to 'converted-to-excel.xlsx' Excel File.")
```
